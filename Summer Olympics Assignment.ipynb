#### 3. Which sport is having most number of medals so far? (Top 5

```python
sport=[]
for sports in df['Sport'].unique():
    sport.append([sports,len(df[df['Sport'] == sports])])
pd.DataFrame(sport,columns = ['Sport','Medals']).sort_values(by='Medals', ascending=False)
```

```python
 pd.DataFrame(sport,columns = ['Sports','Medals']).sort_values(by='Medals', ascending=False).head().plot(x = 'Sports', y = 'Medals', kind = 'bar', figsize = (10,7))
```

##### Aquatics has got most number of all the medals also i.e. 4170 medals

####5. Which player has won most number Gold Medals of medals? (Top 5)

```python
most_gold=[]
gold=df[df['Medal']=='Gold']
for players in gold['Athlete'].unique():
    most_gold.append([players,len(gold[gold['Athlete']==players])])
pd.DataFrame(most_gold,columns = ['Player','Total Gold']).sort_values(by='Total Gold', ascending=False)
```

```python
 pd.DataFrame(most_gold,columns = ['Player','Total Gold']).sort_values(by='Total Gold', ascending=False).head().plot(x = 'Player', y = 'Total Gold', kind = 'bar',color = 'Orange',figsize = (10,7))
```

 ##### PHELPS, Michael has won most number of Gold Medals too

####6. In which year India won first Gold Medal in Summer Olympics?

```python
Ind = df[df['Country']=='IND']
Ind[Ind['Medal']=='Gold'].sort_values(by='Year', ascending=True).head(1)
```

####7. Which event is most popular in terms on number of players? (Top 5)

```python
popular_event=[]
for events in df['Event'].unique():
  popular_event.append([events,len(df[df['Event'] == events])])
pd.DataFrame(popular_event,columns = ['Event','Total Players']).sort_values(by='Total Players', ascending=False)
```

```python
pd.DataFrame(popular_event,columns = ['Event','Total Players']).sort_values(by='Total Players', ascending=False).head().plot(x = 'Event', y = 'Total Players', kind = 'bar',color = 'blue',figsize = (10,7))
```

#####Football is most popular sports in terms of number of players

#### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
female=[]
gold=df[df['Medal']=='Gold']
female_gold = gold[gold['Gender']=='Women']
for sports in female_gold['Sport'].unique():
    female.append([sports,len(female_gold[female_gold['Sport'] == sports])])
pd.DataFrame(female,columns = ['Sport','Total women']).sort_values(by='Total women', ascending=False)
```

```python
 pd.DataFrame(female,columns = ['Sport','Total women']).sort_values(by='Total women', ascending=False).head().plot(x = 'Sport', y = 'Total women', kind = 'bar',color = 'g',figsize = (10,7))
```
